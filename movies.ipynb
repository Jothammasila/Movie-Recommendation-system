```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

```python
# Load the dataset

movies = pd.read_csv('/home/jothammasila/Projects/Datasets/movies.c')
```

```python
movies.head()
```

```python
movies.info()
```

```python
# Select the relevant features for recommendation
selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
# Replacing the null values with null strings

for feature in selected_features:
    movies[feature] = movies[feature].fillna("")
```

```python
# Combining all the 5 selected features
combined_features = movies['genres']+''+movies['keywords']+''+movies['tagline']+''+movies['cast']+''+movies['director']
```

```python
print(combined_features)
```

```python
#Converting text data to feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

## Cosine Similarity

##### Cosine similarity measures the similarity between two vectors of an inner product space. It is measured by the cosine of the angle between two vectors and determines whether two vectors are pointing in roughly the same direction. It is often used to measure document similarity in text analysis.

```python
# Getting the similarity scores using cosine_similarity
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
# similariry.shape[0] is the number. i.e len of the movie dataset
# similariry.shape[1] are the features where each movie is made 
# a feature to be compared with all other movies
print(similarity.shape)
```

```python
# Getting movie name from user
movie_name = input("Enter favourite movie: ")
```

```python
# Creating a list with all the movies given in the dataset
# This list is the one to be compared with the movie given by the user.
list_of_all_titles = movies['title'].tolist()
print(list_of_all_titles)
```

```python
# Finding a close match for movie name given by the user
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
```

```python
print(find_close_match)
```

```python
close_match = find_close_match[0]
```

```python
# Finding the index of the movie with title
index_of_the_movie = movies[movies.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# Getting a list of similar movies
# The list contains the index of a particular movie in the dataset 
# and its similarity_score to the user-preferred movie e.g, the first
# movie in the dataset which is index 0 is 'Avatar' which has a similarity
# score of 0.0 compared when compared to user preferred movie
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on similarity_score

sorted_similar_movies = sorted(similarity_score,
                               key=lambda x:x[1],reverse=True)
# x[1] is score
print(sorted_similar_movies)
```

```python
# Print the name of a similar movie based on the title
print('You would like.\n')

k = 1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies[movies.index == index]['title'].values[0]
    if k<11:
        print(k,' : ',title_from_index)
        k+=1
```

## Putting Everything together.
#### Movie Recommendation System

```python
def recommendation():
    # Getting movie name from user
    movie_name = input("Enter favourite movie: ")
    
    # Creating a list with all the movies given in the dataset
    # This list is the one to be compared with the movie given by the user.
    list_of_all_titles = movies['title'].tolist()
    
    # Finding a close match for movie name given by the user
    find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
    
    close_match = find_close_match[0]
    
    
    # Finding the index of the movie with title
    index_of_the_movie = movies[movies.title == close_match]['index'].values[0]
    
    
    # Getting a list of similar movies
    # The list contains the index of a particular movie in the dataset 
    # and its similarity_score to the user-preferred movie e.g, the first
    # movie in the dataset which is index 0 is 'Avatar' which has a similarity
    # score of 0.0 compared when compared to user preferred movie
    similarity_score = list(enumerate(similarity[index_of_the_movie]))
    
    # sorting the movies based on similarity_score
    
    sorted_similar_movies = sorted(similarity_score,
                                   key=lambda x:x[1],reverse=True)
    # x[1] is score
    
    # Print the name of a similar movie based on the title
    print('You would like.\n')
    
    k = 1
    for movie in sorted_similar_movies:
        index = movie[0]
        title_from_index = movies[movies.index == index]['title'].values[0]
        if k<11:
            print(k,' : ',title_from_index)
            k+=1

recommendation()
```
